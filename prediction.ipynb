## Scaling

```python
def Standard_Scaler(train_data, val_data, test_data):

  scaler = preprocessing.StandardScaler().fit(train_data)
  train_data = scaler.transform(train_data)
  val_data = scaler.transform(val_data)
  test_data = scaler.transform(test_data)
  
  return train_data, val_data, test_data
```

```python
def Min_Max_Scaler(train_data, val_data, test_data):
  
  scaler = preprocessing.MinMaxScaler().fit(train_data)
  train_data = scaler.transform(train_data)
  val_data = scaler.transform(val_data)
  test_data = scaler.transform(test_data)

  return train_data, val_data, test_data
```

## Evaluating method

```python
# https://scikit-learn.org/stable/modules/classes.html#module-sklearn.metrics
BEST_DATASET = ""
BEST_CLASSIFIER = ""
GLOBAL_MSE = 10e100
GLOBAL_MAE = 10e100
GLOBAL_RMSE = 10e100
GLOBAL_R2 = 10e100



def reg_metrics(real, pred, dataset, classifier):
    
    MSE = mean_squared_error(real, pred)
    RMSE = mean_squared_error(real, pred, squared=False)
    MAE = mean_absolute_error(real, pred)
    R2 = r2_score(real, pred)

    print('Evaluating on following')
    print("""MAE = {:.3f}, MSE = {:.3f}, RMSE = {:.3f}",
            R2 = {:.3f}""".format(MAE, MSE, RMSE, R2))
    global GLOBAL_MSE; 
    global GLOBAL_MAE 
    global GLOBAL_RMSE
    global GLOBAL_R2 
    global BEST_DATASET
    global BEST_CLASSIFIER

    if (MSE < GLOBAL_MSE and dataset!=" "):
        print(GLOBAL_MSE)
        GLOBAL_MSE = MSE
        GLOBAL_MAE = MAE
        GLOBAL_RMSE = RMSE
        GLOBAL_R2 = R2
        BEST_DATASET= dataset
        BEST_CLASSIFIER = classifier
        print("***********************************")
        print("""We got a better classifier on {} which is {} with MAE = {:.3f}, MSE = {:.3f}, RMSE = {:.3f}",
            R2 = {:.3f}""".format(dataset, classifier, MAE, MSE, RMSE, R2))
        print("***********************************")

    return MAE, MSE, RMSE, R2


```

### **Classifier-1. LinearRegression**

```python
def LinearR(dataset, trainx, trainy, valx, valy, testx, testy):

  lr = LinearRegression()
  lr.fit(trainx, trainy)
  predictionT = lr.predict(trainx)
  print("Training")
  reg_metrics(trainy, predictionT, " ", "LinearRegression")

  predictionV = lr.predict(valx)
  print("Val")
  reg_metrics(valy, predictionV, " ", "LinearRegression")

  predictionTE = lr.predict(testx)
  print("Test")
  reg_metrics(testy, predictionTE, dataset, "LinearRegression")
```

### **Classifier-3. SVM**

```python
kernel = ''
best_degreeSVR = 0
best_accSVR = 10000000.0
MAE0, MSE0, RMSE0, R20 = 0.0, 0.0, 0.0, 0.0
MAE, MSE, RMSE, R2 = 0.0, 0.0, 0.0, 0.0


def SVMR(dataset, trainx, trainy, valx, valy, testx, testy):

  degrees_list = [1, 2, 3, 4]
  kernels = ['linear', 'rbf', 'sigmoid', 'poly']

  for classifierIn in kernels:

   if classifierIn == 'poly':
    for degree_val in degrees_list:
      svr = SVR(kernel='poly', degree=degree_val)
      svr.fit(trainx, trainy)
      
      # evaluate on X_train_poly, y_train and X_val_poly, y_val
      y_pred_on_train = svr.predict(trainx)
      y_pred_on_val = svr.predict(valx)
      MSE = mean_squared_error(valy, y_pred_on_val)
      RMSE = mean_squared_error(valy, y_pred_on_val, squared=False)
      MAE = mean_absolute_error(valy, y_pred_on_val)
      R2 = r2_score(valy, y_pred_on_val)

      # print("classifier={}, Setting degree={}, train_acc={:.2f}, val_acc={:.2f}".format(classifierIn, degree_val, train_acc, val_acc))
      if RMSE < best_accSVR:
          print("It is better than what we got, let's see the results")
          print("the kernel is",classifierIn)
          print("The degree is ", degree_val)
          print("Traing evaluate: ")
          reg_metrics(trainy, y_pred_on_train, " ", "SVM")
          print("Validation evaluate: ")
          reg_metrics(valy, y_pred_on_val, " ", "SVM")
          print('---------------')
          kernel = classifierIn
          best_degreeSVR = degree_val
          MAE0, MSE0, RMSE0, R20 = MAE, MSE, RMSE, R2

   else:
      svr = SVR(kernel=classifierIn)
      svr.fit(trainx, trainy)
      
      # evaluate on X_train
      y_pred_on_train = svr.predict(trainx)
      y_pred_on_val = svr.predict(valx)
      MSE = mean_squared_error(valy, y_pred_on_val)
      RMSE = mean_squared_error(valy, y_pred_on_val, squared=False)
      MAE = mean_absolute_error(valy, y_pred_on_val)
      R2 = r2_score(valy, y_pred_on_val)
      

      
      # print("classifier={}, Setting degree={}, train_acc={:.2f}, val_acc={:.2f}".format(classifierIn, degree_val, train_acc, val_acc))
      if RMSE < best_accSVR:
          print("It is better than what we got, let's see the results")
          print("the kernel is",classifierIn)
          print("Traing evaluate: ")
          reg_metrics(trainy, y_pred_on_train, " ", "SVM")
          print("Validation evaluate: ")
          reg_metrics(valy, y_pred_on_val, " ", "SVM")
          print('---------------')
          kernel = classifierIn
          MAE0, MSE0, RMSE0, R20 = MAE, MSE, RMSE, R2  


  X_train_val_merge = np.vstack([trainx, valx]) 
  y_train_val_merge = np.concatenate((trainy, valy), axis=0)

  # svclassifier = None

  if kernel == 'poly':
     svrlassifier = SVR(kernel='poly', degree=best_degreeSVR)

  else:
     svrlassifier = SVR(kernel=kernel)
    
  svrlassifier.fit(X_train_val_merge, y_train_val_merge)





  print('---------------')
  print('---------------')
  print('---------------')
  print()
  if kernel == 'poly':    
    print("Best classifier={}, Best setting degree={}, MAE, MSE, RMSE, R2".format(kernel, best_degreeSVR, MAE0, MSE0, RMSE0, R20))

  else:
    print("Best classifier={}, MAE, MSE, RMSE, R2".format(kernel, MAE0, MSE0, RMSE0, R20))

  print('---------------')
  # predict on X_test_poly, y_test
  y_pred_on_test = svrlassifier.predict(testx)
  print("Test Metrics")
  reg_metrics(testy, y_pred_on_test, dataset, "SVM")


```

### **Classifier-4. Lasso**

```python
def LASOO(dataset, trainx, trainy, valx, valy, testx, testy):
  lassoReg = Lasso(alpha=100)
  lassoReg.fit(trainx, trainy)
  # Predict

  y_pred_on_train = lassoReg.predict(trainx)
  print("Training")
  reg_metrics(trainy, y_pred_on_train, " ", "Lasso")
  

  print("Val")
  y_pred_on_val = lassoReg.predict(valx)
  MAE, MSE, RMSE, R2 = reg_metrics(valy, y_pred_on_val, " ", "Lasso")

  print('---------------')


  print("Test")
  y_pred_on_test = lassoReg.predict(testx)
  MAE, MSE, RMSE, R2 = reg_metrics(testy, y_pred_on_test, dataset, "Lasso")
```

### **Classifier-5. ElasticNet**

```python
def ENet(dataset, trainx, trainy, valx, valy, testx, testy):
  ElReg = ElasticNet(alpha=100)
  ElReg.fit(trainx, trainy)
  # Predict

  y_pred_on_train = ElReg.predict(trainx)
  print("Training")
  reg_metrics(trainy, y_pred_on_train, " ", "ENet")
  

  print("Val")
  y_pred_on_val = ElReg.predict(valx)
  MAE, MSE, RMSE, R2 = reg_metrics(valy, y_pred_on_val, " ", "ENet")

  print('---------------')


  print("Test")
  y_pred_on_test = ElReg.predict(testx)
  MAE, MSE, RMSE, R2 = reg_metrics(testy, y_pred_on_test, dataset, "ENet")
```

### **Classifier-6. RandomForestRegressor**

```python
kernel = ''
base_estimators = 0
best_acc = 10000000
MAE0, MSE0, RMSE0, R20 = 0.0, 0.0, 0.0, 0.0

def RandomF(dataset, trainx, trainy, valx, valy, testx, testy):

  MAE, MSE, RMSE, R2 = 0.0, 0.0, 0.0, 0.0

  for i in range(10, 45,5):  
    ran_model = RandomForestRegressor(n_estimators = i,criterion='mse', random_state = 42)
    ran_model=ran_model.fit(trainx,trainy)

    predictionT = ran_model.predict(trainx)
    predictionV = ran_model.predict(valx) 
    MSE = mean_squared_error(valy, predictionV)
    RMSE = mean_squared_error(valy, predictionV, squared=False)
    MAE = mean_absolute_error(valy, predictionV)
    R2 = r2_score(valy, predictionV)



    if RMSE < best_acc:
          print("It is better than what we got, let's see the results")
          print("the kernel is squared_error")
          print("Traing evaluate: ")
          reg_metrics(trainy, predictionT, " ", "RF")
          print("Validation evaluate: ")
          reg_metrics(valy, predictionV, " ", "RF")
          print('---------------')
          kernel = 'mse'
          MAE0, MSE0, RMSE0, R20 = MAE, MSE, RMSE, R2  
          base_estimators = i


  for i in range(10, 45,5):  
    ran_model = RandomForestRegressor(n_estimators = i,criterion='mae', random_state = 42)
    ran_model=ran_model.fit(trainx,trainy)

    predictionT = ran_model.predict(trainx)
    predictionV = ran_model.predict(valx) 
    MSE = mean_squared_error(valy, predictionV)
    RMSE = mean_squared_error(valy, predictionV, squared=False)
    MAE = mean_absolute_error(valy, predictionV)
    R2 = r2_score(valy, predictionV)


    if RMSE < best_acc:
          print("It is better than what we got, let's see the results")
          print("the kernel is absolute_error")
          print("Traing evaluate: ")
          reg_metrics(trainy, predictionT, " ", "RF")
          print("Validation evaluate: ")
          reg_metrics(valy, predictionV, " ", "RF")
          print('---------------')
          kernel = 'mae'
          MAE0, MSE0, RMSE0, R20 = MAE, MSE, RMSE, R2  
          base_estimators = i



  print('---------------')
  print('---------------')
  print('---------------')

  print("Best classifier={}, Best base_estimators={}, MAE, MSE, RMSE, R2".format(kernel, base_estimators, MAE0, MSE0, RMSE0, R20)) 
  print()
  print("Test Metrics")
  ran_model = RandomForestRegressor(n_estimators = base_estimators,criterion=kernel, random_state = 42)
  ran_model=ran_model.fit(trainx,trainy)
  predictionTe = ran_model.predict(testx) 
  reg_metrics(testy, predictionTe, dataset, "RF")
```
